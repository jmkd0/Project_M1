## Preprocessing

Question II

```python
data.isnull().sum()
# Replace NaN value with the mean
data['Age'].fillna(data['Age'].mean(),inplace=True)
#Replace NaN value with the most common Cabin (Max)
data['Cabin'].fillna(data['Cabin'].value_counts().index[0], inplace=True)
#Replace NaN value with the most common Embarked (Max)
data['Embarked'].fillna(data['Embarked'].value_counts().index[0], inplace=True)
data.head()
```

# Visualization

Question III: 2

1-Relation between sex and Survival

```python
sex_survey = data[['Sex','Survived']]
survey = sex_survey[sex_survey['Survived'] == 1].groupby('Sex').count()
dead = sex_survey[sex_survey['Survived'] == 0].groupby('Sex').count()
sexs = survey.index.values.tolist()

labels = ['Survey', 'Dead']
plt.bar(sexs, survey['Survived'])
plt.bar(sexs, dead['Survived'], bottom = survey['Survived'])
plt.title('Relation between sex and Survival\n', loc='left')
plt.xlabel('Sex')
plt.ylabel('Number of people')
plt.legend(labels)
```

2-Relation between Age and Survival

```python
ages = pd.DataFrame(list([i for i in range(100)]))
ages.columns = ['age']

age_and_survey = data[['Age','Survived']]
survey = age_and_survey[age_and_survey['Survived'] == 1].groupby('Age').count()
dead = age_and_survey[age_and_survey['Survived'] == 0].groupby('Age').count()

survey = survey.reset_index()
survey['Age'] = survey['Age'].round()
survey = survey.groupby('Age').sum()
survey = survey.reset_index()
survey = pd.merge(ages,survey, left_on='age', right_on='Age', how='left').fillna(0)
survey = survey[['age', 'Survived']]

dead = dead.reset_index()
dead['Age'] = dead['Age'].round()
dead = dead.groupby('Age').sum()
dead = dead.reset_index()
dead = pd.merge(ages,dead, left_on='age', right_on='Age', how='left').fillna(0)
dead = dead[['age', 'Survived']]

ages_survey = survey.index.values.tolist()
ages_dead = dead.index.values.tolist()

labels = ['Survey', 'Dead']
plt.figure(figsize=(15, 10))
plt.bar(survey['age'], survey['Survived'])
plt.bar(survey['age'], dead['Survived'], bottom = survey['Survived'])
plt.title('Relation between Age and Survival\n', loc='left')
plt.xlabel('Age')
plt.ylabel('Number of people')
plt.legend(labels)
```

3-Relation between Cabin in suvival

```python
cabin_and_survey = data[['Cabin','Survived']]
cabin_survey = cabin_and_survey[cabin_and_survey['Survived'] == 1].groupby('Cabin').sum()
cabin_survey = cabin_survey.reset_index()

plt.figure(figsize=(15, 10))
plt.bar(cabin_survey['Cabin'], cabin_survey['Survived'])
plt.title('Relation between Cabin and Survival\n', loc='left')
plt.xlabel('Cabin')
plt.ylabel('Number of people')
plt.legend(['survey'])

#Death people
cabin_dead = cabin_and_survey[cabin_and_survey['Survived'] == 0].groupby('Cabin').sum()
cabin_dead = cabin_dead.reset_index()
plt.figure(figsize=(10, 5))
plt.bar(cabin_dead['Cabin'], cabin_dead['Survived'])
plt.title('Relation between Cabin and Survival\n', loc='left')
plt.xlabel('Cabin')
plt.ylabel('Number of people')
plt.legend(['Death'])
```

5-Correlation

```python
def plot_correlation_map( df ):
    corr = df.corr()
    s , ax = plt.subplots( figsize =( 12 , 10 ) )
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    s = sns.heatmap(
    corr,
    cmap = cmap,
    square=True,
    cbar_kws={ 'shrink' : .9 },
    ax=ax,
    annot = True,
    annot_kws = { 'fontsize' : 12 })
```

```python
plot_correlation_map(data)
```

La matrice de corrélation permet de repérer la relation qui lie chaque caractéristique des passagers. 
On remarque la correlation entre Parch (nombre d'enfant ou de parents) et SibSp (nombre de frere/soeur ou époux/épouse) est la plus forte. La survie du passager est aussi lié au prix payé avec une corélation de 0.26.
